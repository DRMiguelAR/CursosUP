## Parte 2 — Técnicas fundamentales

Cuatro operaciones que aparecen constantemente en código de bajo nivel,
sistemas embebidos y programación competitiva:

| Operación | Expresión | Qué hace |
|---|---|---|
| Leer bit `k` | `(x >> k) & 1` | Devuelve 1 si el bit `k` está encendido, 0 si no |
| Encender bit `k` | `x \| (1 << k)` | Pone el bit `k` en 1, sin tocar los demás |
| Apagar bit `k` | `x & ~(1 << k)` | Pone el bit `k` en 0, sin tocar los demás |
| Invertir bit `k` | `x ^ (1 << k)` | Cambia el bit `k`: si era 0 queda 1, y viceversa |

**¿Por qué funcionan?**
- `1 << k` es una máscara con solo el bit `k` encendido.
- OR con esa máscara **fuerza** ese bit a 1.
- AND con el complemento (`~`) **fuerza** ese bit a 0.
- XOR con esa máscara **invierte** solo ese bit.

Demo con `x = 0b10110100` (180) y `k = 1`:

````c++
{
    int x = 0b10110100;  // 180
    cout << "x original       = " << setw(3) << x             << "  ->  " << bin8(x)            << endl;
    cout << "leer bit 1       = " << setw(3) << ((x>>1)&1)    << "  (el bit 1 está apagado)"   << endl;
    cout << "encender bit 1   = " << setw(3) << (x|(1<<1))    << "  ->  " << bin8(x|(1<<1))    << endl;
    cout << "apagar bit 2     = " << setw(3) << (x&~(1<<2))   << "  ->  " << bin8(x&~(1<<2))  << endl;
    cout << "invertir bit 7   = " << setw(3) << (x^(1<<7))    << "  ->  " << bin8(x^(1<<7))   << endl;
}
````

| ¿Potencia de 2? | `n > 0 && !(n & (n-1))` | Verificar en O(1) |
