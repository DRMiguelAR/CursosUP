## Requisitos previos

Ejecuta esta celda antes de empezar:

```c++
#include <iostream>
#include <string>
#include <vector>
#include <tuple>
#include <bitset>
#include <deque>
using namespace std;

// Utilería de verificación: imprime ✓ o ✗ y cuenta aciertos
struct Verificador {
    int ok = 0, total = 0;
    template<typename T>
    void check(T obtenido, T esperado, string descripcion) {
        total++;
        if (obtenido == esperado) {
            cout << "  [OK] " << descripcion << "\n";
            ok++;
        } else {
            cout << "  [FALLO] " << descripcion
                 << "  ->  obtenido: " << obtenido
                 << "  esperado: " << esperado << "\n";
        }
    }
    void resumen() {
        cout << "\n  " << ok << "/" << total << " casos correctos";
        if (ok == total) cout << "  — todos bien, listo para OmegaUP!";
        cout << "\n";
    }
};
```

| 7  | `0111`  | `0010` = 2      | `0101` = 5    |

```c++
// TU CÓDIGO AQUÍ — implementa la función
pair<int,int> separarPorPosicion(int n) {
    int a = 0, b = 0;
    // ...
    return {a, b};
}
```

```c++
// VERIFICACIÓN — ejecuta esta celda para revisar tus casos
{
    Verificador v;
    struct Caso { int n, a, b; };
    vector<Caso> casos = {
        {13, 8,  5},
        { 6, 2,  4},
        { 7, 2,  5},
        { 3, 2,  1},
        { 5, 0,  5},
        {10,10,  0},
        {12, 8,  4},
        {55,34, 21},
    };
    for (auto& c : casos) {
        auto [a, b] = separarPorPosicion(c.n);
        string desc = "n=" + to_string(c.n) + " -> (" + to_string(a) + "," + to_string(b) + ")";
        v.check(make_pair(a, b), make_pair(c.a, c.b), desc);
    }
    v.resumen();
}
```

Cada bit conserva su posición original en `n`.

| n  | binario | 1s (der→izq)                                          | a | b |
|----|---------|------------------------------------------------------|---|---|
| 13 | `1101`  | bit0=1(1°), bit2=1(2°), bit3=1(3°)                  | 9 | 4 |
| 6  | `0110`  | bit1=1(1°), bit2=1(2°)                               | 2 | 4 |
| 7  | `0111`  | bit0=1(1°), bit1=1(2°), bit2=1(3°)                  | 5 | 2 |

```c++
// TU CÓDIGO AQUÍ — implementa la función
pair<int,int> separarPorEnum(int n) {
    int a = 0, b = 0;
    // ...
    return {a, b};
}
```

```c++
// VERIFICACIÓN
{
    Verificador v;
    struct Caso { int n, a, b; };
    vector<Caso> casos = {
        {13, 9,  4},
        { 6, 2,  4},
        { 7, 5,  2},
        { 3, 1,  2},
        { 5, 1,  4},
        {10, 2,  8},
        {12, 4,  8},
    };
    for (auto& c : casos) {
        auto [a, b] = separarPorEnum(c.n);
        string desc = "n=" + to_string(c.n) + " -> (" + to_string(a) + "," + to_string(b) + ")";
        v.check(make_pair(a, b), make_pair(c.a, c.b), desc);
    }
    v.resumen();
}
```

|---|----|-------------|

| 4 | 0  | `APAGADA`   |
| 4 | 47 | `PRENDIDA`  |

```c++
// TU CÓDIGO AQUÍ — implementa la función
string chasquidos(int n, long long c) {
    // ...
    return "APAGADA";
}
```

```c++
// VERIFICACIÓN
{
    Verificador v;
    struct Caso { int n; long long c; string res; };
    vector<Caso> casos = {
        {1,  0,          "APAGADA"},
        {1,  1,          "PRENDIDA"},
        {4,  0,          "APAGADA"},
        {4,  47,         "PRENDIDA"},
        {2,  2,          "PRENDIDA"},
        {2,  4,          "APAGADA"},
        {3,  4,          "PRENDIDA"},
        {3,  8,          "APAGADA"},
        {19, 38797311LL, "PRENDIDA"},
        {21, 38797311LL, "APAGADA"},
    };
    for (auto& c : casos) {
        string res = chasquidos(c.n, c.c);
        string desc = "n=" + to_string(c.n) + " c=" + to_string(c.c) + " -> " + res;
        v.check(res, c.res, desc);
    }
    v.resumen();
}
```

## Programa completo para OmegaUP

Una vez que todos los casos pasan, implementa aquí el programa completo

```c++
// Problema 1 — versión OmegaUP (lee hasta 0)
{
    // Simula la entrada: 13 6 7 0
    vector<int> entrada = {13, 6, 7, 0};
    for (int n : entrada) {
        if (n == 0) break;
        auto [a, b] = separarPorPosicion(n);
        cout << a << " " << b << "\n";
    }
    // Salida esperada:
    // 8 5
    // 2 4
    // 2 5
}
```
