## Parte 1 — Paradoja de los promedios

### 1.1 El problema

Un chiste dice: si un mal estudiante de Cómputo se cambia a Economía, **sube el promedio de inteligencia en ambas facultades**.

Dado el IQ de todos los estudiantes, ¿cuántos estudiantes de Cómputo harían que el chiste fuera verdad?

**Entrada:** número de casos `T`; por caso: `n m` (tamaños de Cómputo y Economía) y luego `n + m` IQs.

**Salida:** por caso, cuántos estudiantes de Cómputo cumplen la condición.

```
Entrada          Salida
1                1

5 5
100 101 102 103 104
98 100 102 99 101
```

### 1.2 Análisis

Sea `avg_c = sum_cs / n` el promedio de Cómputo y `avg_e = sum_ec / m` el de Economía.

Un estudiante con IQ `x` cumple el chiste si:

| Condición | Significado | Forma entera (sin division) |
|-----------|-------------|-----------------------------|
| `x < avg_c` | Al salir de Cómputo, el promedio sube | `x * n < sum_cs` |
| `x > avg_e` | Al entrar a Economía, el promedio sube | `x * m > sum_ec` |

Condición combinada: `avg_e < x < avg_c`.

> **¿Por qué no usar `double`?**  
> Si el promedio es exactamente 101.0 y `x = 101`, la comparación `x < 101.0` con `double` puede dar `false` correctamente, pero con errores de representación podría fallar. Usar enteros es exacto.

**Verificación con el ejemplo:**

```
sum_cs = 510,  avg_c = 102
sum_ec = 500,  avg_e = 100

x=100: 100*5=500 < 510 ✓  pero  100*5=500 > 500 ✗  → No cuenta
x=101: 101*5=505 < 510 ✓  y     101*5=505 > 500 ✓  → Cuenta
x=102: 102*5=510 < 510 ✗                            → No cuenta
```

Resultado: **1** ✓

### 1.3 Implementación

Completa la función `resolver` y prueba con el ejemplo del enunciado:

````c++
{
    // Simula la entrada del problema
    string entrada = R"(
1

5 5
100 101 102 103 104
98 100 102 99 101
)";

    istringstream in(entrada);

    int T;
    in >> T;
    while (T--) {
        int n, m;
        in >> n >> m;
        vector<long long> cs(n), ec(m);
        for (auto& x : cs) in >> x;
        for (auto& x : ec) in >> x;

        long long sum_cs = 0, sum_ec = 0;
        for (auto x : cs) sum_cs += x;
        for (auto x : ec) sum_ec += x;

        int cnt = 0;
        for (auto x : cs)
            if (x * n < sum_cs && x * m > sum_ec)
                cnt++;

        cout << cnt << "\n";
    }
    // Salida esperada: 1
}
````

## Parte 2 — Ayudando a mi hermanito

### 2.1 El problema

**Fuente:** [omegaUp — La-tarea-de-mi-hermanito](https://www.omegaup.com/arena/problem/La-tarea-de-mi-hermanito/)

Mi hermano está aprendiendo a ubicar números de tres dígitos en la recta numérica.
Le doy dos números; él los lee **al revés** y dice cuál es mayor.

**Entrada:** dos números de tres dígitos `A` y `B`, sin ceros, distintos entre sí.

**Salida:** el mayor de los dos **leídos en reversa**, escrito en reversa.

```
Entrada    Salida
735 892    537       (537 > 298)
331 341    143       (133 < 143)
```

### 2.2 Análisis

Para un número de tres dígitos `n = d1 d2 d3` (centenas, decenas, unidades):

```
d1 = n / 100          ← centena
d2 = (n / 10) % 10    ← decena
d3 = n % 10           ← unidad

rev(n) = d3 * 100 + d2 * 10 + d1
       = (n % 10) * 100 + ((n / 10) % 10) * 10 + n / 100
```

Como no hay ceros, la reversa siempre tiene tres dígitos (no aparece un cero inicial).

**Verificación:**
```
rev(735) = 5*100 + 3*10 + 7 = 537
rev(892) = 2*100 + 9*10 + 8 = 298
537 > 298  →  salida: 537 ✓

rev(331) = 1*100 + 3*10 + 3 = 133
rev(341) = 1*100 + 4*10 + 3 = 143
143 > 133  →  salida: 143 ✓
```

### 2.3 Implementación

````c++
{
    auto rev = [](int n) {
        return (n % 10) * 100 + ((n / 10) % 10) * 10 + n / 100;
    };

    // Ejemplo 1: 735 892 → esperado 537
    {
        int A = 735, B = 892;
        int rA = rev(A), rB = rev(B);
        cout << "rev(" << A << ")=" << rA
             << "  rev(" << B << ")=" << rB
             << "  mayor: " << (rA > rB ? rA : rB) << "\n";
    }

    // Ejemplo 2: 331 341 → esperado 143
    {
        int A = 331, B = 341;
        int rA = rev(A), rB = rev(B);
        cout << "rev(" << A << ")=" << rA
             << "  rev(" << B << ")=" << rB
             << "  mayor: " << (rA > rB ? rA : rB) << "\n";
    }
}
````
