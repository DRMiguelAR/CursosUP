## 1. ¿Qué es C++ y por qué aprenderlo?

C++ es un lenguaje de programación de **propósito general** creado por [Bjarne Stroustrup](https://www.stroustrup.com/) a principios de los años 80 como extensión de C. Su nombre alude al operador de incremento (`++`): *C mejorado*.

### Características principales

| Característica | Significado práctico |
|---|---|
| **Compilado** | El código se traduce a lenguaje máquina antes de ejecutarse → muy rápido |
| **Tipado estático** | Los tipos de las variables se verifican en compilación → menos errores en tiempo de ejecución |
| **Acceso a memoria** | Control directo sobre la RAM mediante apuntadores |
| **Multi-paradigma** | Soporta programación estructurada, orientada a objetos y genérica |
| **Estándar vivo** | Se actualiza periódicamente: C++11, C++14, C++17, C++20, C++23 |

### ¿Dónde se usa?

- Sistemas operativos (Windows, partes de Linux/macOS)
- Motores de videojuegos (Unreal Engine, Unity internamente)
- Bases de datos (MySQL, MongoDB)
- Aplicaciones de alto rendimiento: navegadores (Chrome, Firefox), compiladores, simuladores físicos
- Sistemas embebidos y robótica

## 3. Compiladores actuales

Un **compilador** traduce tu código fuente (`.cpp`) a código máquina (`.exe` o binario ejecutable). Los tres más relevantes hoy en día son:

### g++ — GNU Compiler Collection
- El compilador de referencia en Linux y macOS
- Disponible en Windows a través de [MinGW](https://www.mingw-w64.org/) o [WSL](https://learn.microsoft.com/en-us/windows/wsl/)
- Comando: `g++`
- Documentación: https://gcc.gnu.org/

### clang++ — LLVM Compiler Infrastructure
- Desarrollado principalmente por Apple, ampliamente usado en macOS
- Mensajes de error más claros y descriptivos que g++
- Comando: `clang++`
- Documentación: https://clang.llvm.org/

### cling — C++ Interpreter
- **No es un compilador tradicional**: es un intérprete interactivo basado en LLVM/Clang
- Permite ejecutar C++ línea a línea, sin necesidad de compilar un archivo completo
- Es el motor que usa **xeus-cling**, el kernel de C++ para Jupyter — es decir, el que estamos usando en este cuaderno
- Documentación: https://root.cern/cling/

> **¿Por qué importa la diferencia?**  
> `g++` y `clang++` compilan: generan un ejecutable completo. `cling` interpreta: ejecuta celda a celda, como Python. Esto lo hace ideal para aprender de forma interactiva, pero en producción siempre se usa un compilador real.

## 4. Entornos de trabajo

### 4.1 Consola / Terminal
El entorno más básico y universal. Compilas con `g++` o `clang++` y ejecutas el binario resultante. Funciona en Linux, macOS y Windows (con WSL o MinGW). No requiere ninguna instalación adicional más allá del compilador.

### 4.2 Visual Studio Code
Editor de código ligero y extensible. Con las extensiones adecuadas ofrece resaltado de sintaxis, autocompletado, depuración y terminal integrada.

- Descarga: https://code.visualstudio.com/
- Extensión recomendada: [C/C++ (Microsoft)](https://marketplace.visualstudio.com/items?itemName=ms-vscode.cpptools)
- Guía de configuración para C++: https://code.visualstudio.com/docs/languages/cpp

### 4.3 Xeus-cling en Jupyter (este cuaderno)
Combina la interactividad de Jupyter con C++. Permite ejecutar código celda a celda, ver resultados inmediatos y mezclar explicaciones (Markdown) con código. Ideal para aprender y documentar.

- Proyecto xeus-cling: https://github.com/jupyter-xeus/xeus-cling
- Limitación: no soporta entrada interactiva por teclado (`cin` no funciona igual que en consola)

### 4.4 Google Cloud Shell
Terminal Linux completa en el navegador, sin instalar nada. Incluye `g++` preinstalado. Útil para practicar desde cualquier equipo.

- Acceso: https://console.cloud.google.com → ícono de terminal (arriba a la derecha)
- O directamente: https://shell.cloud.google.com

## 5. Compilar y ejecutar desde la consola con `g++`

### Flujo básico

```
código fuente          compilación            ejecución
 main.cpp      →    g++ main.cpp -o hola   →   ./hola
 (texto)             (genera binario)          (corre el programa)
```

### Comandos paso a paso

**1. Verificar que g++ está instalado:**
```bash
g++ --version
```

**2. Compilar un archivo:**
```bash
g++ main.cpp -o hola
```
- `main.cpp` — archivo fuente
- `-o hola` — nombre del ejecutable de salida (`-o` de *output*)

**3. Ejecutar:**
```bash
./hola
```

**4. Compilar especificando el estándar (recomendado):**
```bash
g++ -std=c++17 main.cpp -o hola
```

**5. Compilar con avisos activados (buena práctica):**
```bash
g++ -std=c++17 -Wall -Wextra main.cpp -o hola
```
- `-Wall` — activa los avisos más comunes
- `-Wextra` — activa avisos adicionales

### ¿Qué pasa si hay errores?

Si el código tiene errores de sintaxis o tipo, g++ **no genera el ejecutable** y en su lugar imprime mensajes de error indicando el archivo, la línea y la descripción del problema:

```
main.cpp:5:5: error: expected ';' after return statement
     return 0
            ^
```

Aprende a leer estos mensajes: archivo → número de línea → descripción.

## 6. Primer programa: estructura básica

Este es el archivo `main.cpp` que usamos en clase:

```cpp
#include <iostream>
using namespace std;

int main(){
    cout << "Hello World";
}
```

### Anatomía del programa

| Línea | Qué hace |
|---|---|
| `#include <iostream>` | **Directiva de preprocesador.** Incluye la biblioteca de entrada/salida estándar. Sin esta línea, `cout` no existe. |
| `using namespace std;` | Permite usar `cout`, `cin`, `endl`, etc. directamente. Sin ella habría que escribir `std::cout`. |
| `int main()` | **Punto de entrada del programa.** Todo programa C++ debe tener exactamente una función `main`. El `int` indica que devuelve un entero al sistema operativo. |
| `{ ... }` | Cuerpo de la función. El código que se ejecuta está entre llaves. |
| `cout << "Hello World";` | Imprime texto en la consola. `<<` es el operador de inserción de flujo. |

> **Nota sobre `return`:** En `main`, el `return 0;` es opcional en C++11 en adelante. Si se omite, el compilador asume que el programa terminó correctamente. En otras funciones sí es obligatorio si el tipo de retorno no es `void`.

### Ejecutemos el mismo programa en este cuaderno

> En xeus-cling no necesitamos la función `main` — el código se ejecuta directamente. Tampoco repetimos los `#include` si ya los ejecutamos antes.

````c++17
#include <iostream>
using namespace std;
````

````c++17
{
    cout << "Hello World" << endl;
}
````

**¿Por qué funciona sin `main`?**  
xeus-cling (a través de cling) mantiene un contexto de ejecución persistente. No necesita una función `main` porque ya tiene su propio entorno de evaluación, igual que el intérprete de Python. Las variables y funciones que defines en una celda quedan disponibles para las siguientes.

### Variaciones de `cout`

`cout` puede encadenar múltiples valores con `<<` y usar `endl` para saltar de línea:

````c++17
{
    cout << "Hola" << endl;                     // endl: salto de línea + vacía el buffer
    cout << "Línea 1" << "\n" << "Línea 2\n";   // \n: solo salto de línea
    cout << "Valor: " << 42 << endl;            // se pueden mezclar texto y números
    cout << "Pi ~= " << 3.14159 << endl;
}
````

> **`endl` vs `\n`:** Ambos generan un salto de línea, pero `endl` además **vacía el buffer de salida** (flush). En la mayoría de los casos `\n` es suficiente y más rápido. Usa `endl` cuando necesites asegurarte de que la salida se muestra inmediatamente.

## 7. Entrada y salida: `cin`

Este es el archivo `main3.cpp` de clase:

```cpp
#include <iostream>
#include <string>
using namespace std;

int main(){
    string nombre;
    cout << "Introduce tu nombre: ";
    cin >> nombre;
    cout << "Hello " << nombre << endl;
}
```

### `cin` — entrada estándar

| Elemento | Descripción |
|---|---|
| `#include <string>` | Necesario para usar el tipo `string` |
| `string nombre;` | Declara una variable de texto |
| `cin >> nombre;` | Lee una palabra del teclado y la guarda en `nombre` |
| `>>` | Operador de extracción de flujo (opuesto a `<<`) |

### En consola se compila y ejecuta normalmente:

```bash
g++ -std=c++17 main3.cpp -o saludo
./saludo
```
```
Introduce tu nombre: Ana
Hello Ana
```

### En el cuaderno: `cin` no es interactivo

xeus-cling no puede esperar input del teclado durante la ejecución de una celda. Para demostrar el mismo concepto en el cuaderno, asignamos el valor directamente a la variable:

````c++17
#include <string>
````

````c++17
{
    string nombre = "Ana";   // en consola esto vendría de: cin >> nombre;
    cout << "Hello " << nombre << endl;
}
````

## 8. Comparación: consola vs cuaderno

| Aspecto | Consola con g++ | Cuaderno (xeus-cling) |
|---|---|---|
| Necesita `main()` | Sí, siempre | No |
| Necesita `#include` | Sí, en cada archivo | Sí, pero solo una vez por sesión |
| `cin` interactivo | Sí | No |
| Redeclarar variables | Error | Posible si se usan bloques `{}` |
| Genera ejecutable | Sí (`.exe` o binario) | No, evalúa en memoria |
| Útil para | Programas completos, producción | Aprendizaje, experimentación, documentación |
